### Task 1
Write a program that stores the information of a Bank Account (Account class) and then it allows the user to deposit, withdraw and display balance. The class should include the following variables:  
- name
- balance (private)
  
Use the following guidelines:
- In Account class:
    - Add a constructor. 
    - Add getters for private variables.
    - Create function deposit that receives amount and update the balance after deposit.
    - Create function withdraw that receives amount and update the balance after withdrawing.
    - Add a Destructor.
      
- Ask the User for input:
    - Enter name and id, then create an object.
    - Choose the Option:
        1. deposit
        2. withdraw
        3. Display balance
        4. Exit
    - Perform the Operation
    - Repeat until Option 4 is chosen
    - After the user exit menu , add VIP attribute if balance is greater than 10,000 and print ‎the account information.‎
    - Destroy the object


| sample output 1 | sample output 2 |
|-----------------|-----------------|
|  |  |

```python
class Account:
    def __init__(self, name, balance=0.0):
        self.__balance = balance   # private variable
        self.name = name
        print(f"{self.name} Acount Created")

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def display_balance(self):
        print(f"Your Balance is: {self.__balance:.0f}")

    def get_balance(self):
        return self.__balance

    def __del__(self):
        print(f"\n{self.name} Acount Destroyed")


print("Welcome to AI Bank")

name = input("Enter your name:")
initial_balance = float(input("Enter your initial balance:"))

acc = Account(name, initial_balance)

while True:
    print("\n1 - deposit")
    print("2 - withdraw")
    print("3 - Display balance")
    print("4 - Exit")

    choice = int(input("Choose Operation: "))

    if choice == 1:
        amount = float(input("Enter amount to deposit"))
        acc.deposit(amount)

    elif choice == 2:
        amount = float(input("Enter amount to withdraw"))
        acc.withdraw(amount)

    elif choice == 3:
        acc.display_balance()

    elif choice == 4:
        break

# After exit
balance = acc.get_balance()
if balance > 10000:
    print("Your are  VIP")
else:
    print("Your are  not VIP")

print(f"Your Balance is: {balance:.0f}")

del acc
```
