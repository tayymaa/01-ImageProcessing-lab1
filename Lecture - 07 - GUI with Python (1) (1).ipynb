### Task 1 : GUESSING NUMBER 
Write a program that randomly generates an integer between 0 and 10. The program prompts the user to enter a number. When the user click the button a message should show:
- “you win “ if random number=input
- “too high “ if random number<input
- “you low “ if random number>input






| ouput 1 | ouput 2 | ouput 3 |
|----------|----------|----------|
|  | | |

```python
import tkinter as tk
import random

window = tk.Tk()
window.title("first gui :)")

r = random.randint(0,10)
tk.Label(window, text="guess a magic number between 0 and 10 !").pack()
guess = tk.Entry(window, font=("Arial", 12))
guess.pack(padx=10, pady=5)
msg = tk.Label(window, text="")
msg.pack(pady=5)

def check():
    try:
        g = int(guess.get())
        if g == r:
            msg.config(text="you win!")
        elif g > r:
            msg.config(text="too high!")
        else:
            msg.config(text="too low!")
    except:
        msg.config(text="enter a number")

btn = tk.Button(window, text="submit", command=check)
btn.pack(pady=5)

window.mainloop()
```

### Task 2 : CALCULATOR MENU
Write a program that has main menu: calculator , in calculator the  sub menu:
- ADD: will show button “add “ to add the two numbers.
- SUB: will show button “sub “ to subtract the two numbers.
- MUL: will show button “mul “ to multiply the two numbers.
- DIV: will show button “div “ to divide the two numbers.

<center></center>

```python
import tkinter as tk

window = tk.Tk()
window.title("Calculator")

first_number = tk.Entry(window, font=("Arial", 12))
second_number = tk.Entry(window, font=("Arial", 12))
first_number.pack(fill="x", padx=12, pady=(10, 6))
second_number.pack(fill="x", padx=12, pady=(0, 10))
first_number.insert(0, "10")
second_number.insert(0, "5")
area = tk.Frame(window)
area.pack(pady=4)

def get_numbers():
    try:
        a = float(first_number.get())
        b = float(second_number.get())
        return a, b, ""
    except:
        return 0.0, 0.0, "please enter valid numbers"

def show(text):
    tk.Label(area, text=text, font=("Arial", 11)).pack(pady=2)

def add_calc_button(name, op):
    def run():
        a, b, err = get_numbers()
        if err:
            show(err)
            return
        if op == "+":
            r = a + b
        elif op == "-":
            r = a - b
        elif op == "*":
            r = a * b
        elif op == "/":
            if b == 0:
                show("can’t divide by zero")
                return
            r = a / b
        show(f"result={r}")
    tk.Button(area, text=name.lower(), command=run).pack(pady=3)

menu_bar = tk.Menu(window)
menu_calc = tk.Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="Calculator", menu=menu_calc)
menu_calc.add_command(label="ADD", command=lambda: add_calc_button("add", "+"))
menu_calc.add_command(label="SUB", command=lambda: add_calc_button("sub", "-"))
menu_calc.add_command(label="MUL", command=lambda: add_calc_button("mul", "*"))
menu_calc.add_command(label="DIV", command=lambda: add_calc_button("div", "/"))
window.config(menu=menu_bar)
window.mainloop()
```

### Task 3: BOOK A TICKET
Write a program for booking a new ticket. Ask the user about :
- Name
- Destination
- Flight Time
- Drinks
After submission show the ticket information in green color.

*use widgets as in the sample output.*

<center></center>


```python
import tkinter as tk
from tkinter import ttk

window = tk.Tk()
window.title("New Ticket")
window.configure(bg="black")

name = tk.Entry(window, font=("Arial", 12))
name.insert(0, "Tayma") 
name.pack(padx=10, pady=5, fill="x")

place = tk.StringVar(value="Kuwait")
places = ttk.Combobox(window, textvariable=place, values=["Kuwait","Riyadh","Jeddah","Dammam","Doha"])
places.pack(padx=10, pady=5)

tk.Label(window, text="preferred flight time", bg="black", fg="white", font=("Arial", 11)).pack()
time = tk.StringVar(value="Evening")
tk.Radiobutton(window, text="Morning", variable=time, value="Morning", bg="black", fg="white").pack()
tk.Radiobutton(window, text="Evening", variable=time, value="Evening", bg="black", fg="white").pack()
tk.Label(window, text="preferred drinks on the plane:", bg="black", fg="white", font=("Arial", 11)).pack()

drink1 = tk.IntVar(value=1)
drink2 = tk.IntVar(value=0)
drink3 = tk.IntVar(value=1)
tk.Checkbutton(window, text="coffee", variable=drink1, bg="black", fg="white").pack()
tk.Checkbutton(window, text="tea", variable=drink2, bg="black", fg="white").pack()
tk.Checkbutton(window, text="water", variable=drink3, bg="black", fg="white").pack()
result = tk.Label(window, text="", bg="white", fg="green")
result.pack(padx=10, pady=10, fill="x")

def show_ticket():
    d = []
    if drink1.get(): d.append("coffee")
    if drink2.get(): d.append("tea")
    if drink3.get(): d.append("water")
    txt = "your ticket:\n"
    txt += "name:" + name.get() + "\n"
    txt += "destination:" + place.get() + "\n"
    txt += "flight time:" + time.get() + "\n"
    txt += "drinks:" + ", ".join(d)

    result.config(text=txt)
button = tk.Button(window, text="submit", command=show_ticket)
button.pack()
window.mainloop()
```
