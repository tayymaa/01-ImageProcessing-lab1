### Task 1
Write a program to create a numpy array with students names ['Reem', 'Sara', 'Maha', 'Lama', 'Waad', 'Eman', 'Noha', 'Haya', 'Renad', 'Mona'] and then sort the students’ names and spilt it into two groups each group has 5 students.

Sample Output:
```
group 1: ['Eman' 'Haya' 'Lama' 'Maha' 'Mona']
group 2: ['Noha' 'Reem' 'Renad' 'Sara' 'Waad']


````python
import numpy as np

students = np.array(['Reem', 'Sara', 'Maha', 'Lama', 'Waad', 
                     'Eman', 'Noha', 'Haya', 'Renad', 'Mona'])

students_sorted = np.sort(students)
group1, group2 = np.split(students_sorted, 2)
print("group 1:", group1)
print("group 2:", group2)
````

### Task 2
Create two numpy arrays as follows:  
- students names array: ['Reem', 'Sara', 'Maha', 'Lama', 'Waad', 'Eman', 'Noha', 'Haya', 'Renad', 'Mona']  
- students grades array: [95, 99, 88, 89, 87, 90, 89, 90, 87, 88]

Then write program to print the following:
- students’ names who got above 90 in the homework.
- Highest grade
- Lowest grade
- Randomly pick a student name to solve the homework.
  
You should create two arrays:  
Array 1: students’ names  
Array 2: to hold the students' grades.

Sample Output:
```
A+ students are:
['Sara' 'Reem']
highest grade:  99
lowest grade:  87
Sara will solve the homework
```

````python
import numpy as np
import random

students = np.array(['Reem', 'Sara', 'Maha', 'Lama', 'Waad',
                     'Eman', 'Noha', 'Haya', 'Renad', 'Mona'])
grades = np.array([95, 99, 88, 89, 87, 90, 89, 90, 87, 88])

A_plus_students = students[grades > 90]

highest_grade = np.max(grades)
lowest_grade = np.min(grades)
random_student = random.choice(students)

print("A+ students are:")
print(A_plus_students)
print("highest grade:", highest_grade)
print("lowest grade:", lowest_grade)
print(random_student, "will solve the homework")
````

### Task 3
Write a program that calculate and classify the rainfall average for 4 weeks.
1. Firstly, create 2D array (4,7)  to store the rainfall rates for each day in the 4 weeks.

week#1 |  9 8 6 9 7 6 8  
week#2 | 2 1 2 3 2 3 5  
week#3 | 3 5 5 7 3 2 6  
week#4 | 4 3 5 6 7 3 2  


2. Calculate the rainfall average for each week (i.e. for each row), and print the classification for that week: High, Medium or Low; based on the average as follows: 

| Rainfall average | Classification |
|------------------|----------------|
| average >= 6     |High            |
|6 > average > 3   |Medium          |
|average < = 3     |Low             |



Sample Output:
```
Classification for each week: 

Week  1 :
High : 7.57

Week  2 :
Low : 2.57

Week  3 :
Medium : 4.43

Week  4 :
Medium : 4.29

````python
import numpy as np

rainfall = np.array([
    [9, 8, 6, 9, 7, 6, 8],
    [2, 1, 2, 3, 2, 3, 5],
    [3, 5, 5, 7, 3, 2, 6],
    [4, 3, 5, 6, 7, 3, 2]
])

weekly_avg = np.mean(rainfall, axis=1)
print("Classification for each week:\n")
for i in range(len(weekly_avg)):
    avg = weekly_avg[i]
    if avg >= 6:
        classification = "High"
    elif avg > 3:
        classification = "Medium"
    else:
        classification = "Low"
    
    print(f"Week {i+1}:")
    print(f"{classification} : {avg:.2f}\n")
````

### Task 4
Write a program to remove the negative values in a NumPy array with 0.

Sample Output:
```
Original array:
[-1 -4  0  2  3  4  5 -6]
Replace the negative values of the said array with 0:
[0 0 0 2 3 4 5 0]

````python
import numpy as np

arr = np.array([-1, -4, 0, 2, 3, 4, 5, -6])
print("The original array:")
print(arr)

for i in range(len(arr)):
    if arr[i] < 0:
        arr[i] = 0
print("Replace the negative values of the said array with 0:")
print(arr)
````
